```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_salary = df[pd.notna(df['salary_year_avg'])].copy()

def inflated_salary(salary):
    return salary * 1.03

df_salary['projected_salary'] = df['salary_year_avg'].apply(inflated_salary)

df_salary[['projected_salary', 'salary_year_avg']]
```

```python
df_salary['projected_salary'] = df['salary_year_avg'].apply(lambda salary: salary * 1.03)
df_salary[['projected_salary', 'salary_year_avg']]
```

```python
df_salary['projected_salary'] = df['salary_year_avg'] * 1.03
df_salary[['projected_salary', 'salary_year_avg']]
```

```python
import ast

df.dropna(subset='job_skills', inplace=True)
df.reset_index(drop=True, inplace=True)
df['job_skills'] = df['job_skills'].apply(ast.literal_eval)
```
